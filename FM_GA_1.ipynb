#Import yfinance to Google Sheet

```python
!pip install yahoofinancials
import pandas as pd
from yahoofinancials import YahooFinancials
from dataclasses import dataclass
@dataclass
class FinancialStatements_DATA:
  ticker: str = 'KO'
  freq: str = 'annual'
  stmt: str = 'balance'
  stmt_history: str = 'balanceSheetHistory'

  def get_stmt(m):
    yahoo_financials = YahooFinancials(m.ticker)
    data= yahoo_financials.get_financial_stmts(m.freq, m.stmt)
    dict_list = data[m.stmt_history][m.ticker]
    df = pd.concat([pd.DataFrame(i) for i in dict_list], axis=1)
    df = df.rename_axis(m.stmt_history).reset_index()
    df.insert(0, 'ticker', m.ticker)
    return df

  def get_keystats(m):
    yahoo_financials = YahooFinancials(m.ticker)
    data= yahoo_financials.get_key_statistics_data()
    dict_list = data[m.stmt_history][m.ticker]
    df = pd.concat([pd.DataFrame(i) for i in dict_list], axis=1)
    df = df.rename_axis(m.stmt_history).reset_index()
    df.insert(0, 'ticker', m.ticker)
    return df

  def get_nshares(m):
    yahoo_financials = YahooFinancials(m.ticker)
    data= yahoo_financials.get_num_shares_outstanding(price_type='current')
    return data

  def get_shareprice(m):
    yahoo_financials = YahooFinancials(m.ticker)
    data= yahoo_financials.get_current_price()
    return data
```
